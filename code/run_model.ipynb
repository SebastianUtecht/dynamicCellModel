### Running the simulation

```python
N = 5_00                # Number of cells in the simulation
R = N**(1/3)            # Radius of initial cell distribution
type0_frac    = 1       # Fraction of type 0 cells if using make_random_sphere() cell generation function

# Simulation parameters
sim_dict = {
    # Data and output
    'output_folder'     : "no_gamma_neigh",                                 # Output folder for simulation data
    'data'              : (make_random_sphere ,[N, 1.0-type0_frac, R]),     # Tuple containing data information. Can either be a tuple of form (data_gen, *data_gen_args) to generate the data or tuple (p_mask, x, p, q)
    'yield_every'       : 50,                                               # How often the simulation yields data. 
    'yield_steps'       : 5_00,                                             # How many data-yields we want. Total number of timesteps is yield_every * yield_steps

    # Metasimulational parameters
    'device'            : 'cuda',           # Device to run the simulation on. 'cuda' or 'cpu'
    'dtype'             : torch.float,      # Data type for tensors. Either float32 or float64
    'random_seed'       : 42,               # Random seed for the simulation

    # Proliferation parameters
    'prolif_rate'       : 0.0000,           # Cell division probabilities
    'prolif_delay'      : 0,                # Timesteps before the cells begin proliferating
    'max_cells'         : 10_000,           # Maximum number of cells in the simulation. When this number is reached the simulation terminates

    # Main model parameters
    'lambdas'           : [0., 0.8, 0.2, 0.],       # Lambda parameters for the simulation
    'etas'              : 0.0000,                   # Noise level for the simulation
    'alpha_params'      : [(0.0, 'locked'), (0.0, 'locked')],                         # Alpha parameters for the simulation
    'alpha_range'       : 100,                      # Range for alpha parameters
    'gamma_params'      : [1.2, 'locked'],          # Gamma parameters for the simulation
    'gamma_range'       : 1.5,                      # Range for gamma parameters
    'cell_wall_interaction' : 0.0,                  # Strength of cell-wall interactions. 0 for no interaction, 1 for same strength as cell-cell interactions
    'dt'                : 0.2,                      # Time step for the simulation

    # Boundary parameters
    'bound_radius'      : None,                     # If set, we enact boundary conditions on a spherical region of this radius
    
    # cell distance parameters
    'interaction_dist'  : 3.0,                      # Maximum distance for interactions.
    'offsets'        : [[0, 0], [0, 0]],         

    
    # Miscellaneous
    'notes'             : f'shit_still_fucked_yo',  # Notes about the simulation. Will be printed when the simulation starts if verbose and will also be saved in .json output
    'verbose'           : True,                     # Verbosity level for the simulation. Set off when debugging
    }

run_simulation(sim_dict=sim_dict)                   # Let's run the simulation
```
