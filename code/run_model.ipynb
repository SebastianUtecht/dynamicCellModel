```python
N = 5_00                # Number of cells in the simulation
R = (N/5)**(1/2)            # Radius of initial cell distribution
stretch_frac    = 0.15       # Fraction of type 0 cells if using make_random_sphere() cell generation function


#making a folder for the stuff
folder_name = '07_04_TESTING'
try:
    os.mkdir(folder_name)
except FileExistsError:
    pass

neighbour_type = 'voronoi'
elong_func_type = 'cos'
q_mean = False

eq_lst = ['messy', 'ordered']
data_lst = ['sphere', 'plain']

data_dict = np.load("plain_data/data.pkl.npy", allow_pickle=True).item()
use_trans_neighbors_bool = True

gamma_params = [(1.0, 'free'), (1.0, 'free')]

for data_type in data_lst:
    for eq in eq_lst:
        if data_type == 'plain':
            if eq == 'messy':
                data_dict = (make_stretch_plain ,[30, 0.15])
            elif eq == 'ordered':
                data_dict = np.load("plain_data/data.pkl.npy", allow_pickle=True).item()
        elif data_type == 'sphere':
            if eq == 'messy':
                data_dict = (make_sphere_surface_stretch ,[N, stretch_frac, True, R])
            elif eq == 'ordered':
                data_dict = np.load("start_conf/last_values.pkl.npy", allow_pickle=True).item()
        
        name = data_type + '_' + eq

        # Simulation parameters
        sim_dict = {
            # Data and output
            'output_folder'     : folder_name + f'/{name}',                                 # Output folder for simulation data
            'data'              : data_dict,     # Tuple containing data information. Can either be a tuple of form (data_gen, *data_gen_args) to generate the data or tuple (p_mask, x, p, q)
            'yield_every'       : 50,                                               # How often the simulation yields data. 
            'yield_steps'       : 5_00,                                            # How many data-yields we want. Total number of timesteps is yield_every * yield_steps

            # Metasimulational parameters
            'device'            : 'cuda',           # Device to run the simulation on. 'cuda' or 'cpu'
            'dtype'             : torch.float,      # Data type for tensors. Either float32 or float64
            'random_seed'       : 42,               # Random seed for the simulation

            # Proliferation parameters
            'prolif_rate'       : [0.00000, 0.00000],           # Cell division probabilities
            'prolif_delay'      : 0,                # Timesteps before the cells begin proliferating
            'max_cells'         : 10_000,           # Maximum number of cells in the simulation. When this number is reached the simulation terminates

            # Main model parameters
            'lambdas'           : [[0., 0.8, 0.2, 0.0],
                                    [0., 0.8, 0.2, 0.0],
                                    [0., 0.8, 0.2, 0.0]],                # Lambda parameters for the simulation
            'etas'              :  [0.000, 0.000],                       # Noise level for the simulation
            'alpha_params'      :  [[(0.0, 'locked'), (0.0, 'locked')],
                                    [(0.0, 'locked'), (0.0, 'locked')]],                     # Alpha parameters for the simulation
            'alpha_range'       : 45,                                                       # Range for alpha parameters
            'gamma_params'      : gamma_params,                           # Gamma parameters for the simulation
            'gamma_range'       : 2.01,                      # Range for gamma parameters
            'cell_wall_interaction' : 0.0,                  # Strength of cell-wall interactions. 0 for no interaction, 1 for same strength as cell-cell interactions
            'dt'                : 0.2,                      # Time step for the simulation

            # Boundary parameters                                                                                                                                                                                                                                     
            'bound_type'        : None,
            'bound_extents'     : [24, 4],
            'bound_move_times'  : [3_000, 17_000],
            'bound_continuity'  : 5_000,

            # Stretching parameters
            'stretch_factor'     : 0.0, #0.01,
            
            # cell distance parameters-
            'interaction_dist'  : 4.0,                              # Maximum distance for interactions.
            'offsets'        : [[0, 0], [0, 0]],

            #neighbor debugging params
            'neighbour_type'    : neighbour_type,                           # Whether to use voronoi or knear neighbors
            'use_q_mean'        : q_mean,                                   # Whether to use the mean of q_i and q_j for interactions or just q_i
            'elong_func_type'   : elong_func_type,                          # Whether to use cos(2theta) or linear function for elongation. 'cos' or 'linear'
            'use_trans_neighbors': use_trans_neighbors_bool,

            # Miscellaneous
            'notes'             : name,  # Notes about the simulation. Will be printed when the simulation starts if verbose and will also be saved in .json output
            'verbose'           : True,                                                                     # Verbosity level for the simulation. Set off when debugging
            }

        try:
            run_simulation(sim_dict=sim_dict)                                                               # Let's run the simulation
        except Exception as e:
            print(f"Error occurred while running simulation: {e}")
```
